```python
import seaborn as sns
%matplotlib inline
```

Heatmap

```python
import numpy as np
import matplotlib.pyplot as plt
#Generating 2-D 10x10 matrix of random numbers
#from 1 to 100
data=np.random.randint(low=1,high=100,size=(10,10))
print("The data to be plotted:\n")
print(data)
#plotting the heatmap
hm=sns.heatmap(data=data)
#displaying the plotted heatmap
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print("The data to be plotted:\n")
print(data)
# setting the parameter values
annot=True
#plotting the heatmap
hm=sns.heatmap(data=data,annot=annot)
#displaying the plotted heatmap
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
cmap="tab10"
print("The data to be plotted:\n")
print(data)
#plotting the heatmap
annot=True
hm=sns.heatmap(data=data,cmap=cmap,annot=annot)
#displaying the plotted heatmap
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print("The data to be plotted:\n")
print(data)
#setting the parameters values
linewidths=2
linecolor="yellow"
annot=True
#plotting the heatmap
hm=sns.heatmap(data=data,linewidths=linewidths,linecolor=linecolor,annot=annot)
#displaying the plotted heatmap
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print("The data to be plotted:\n")
print(data)
#setting the parameters values
linewidths=2
linecolor="yellow"
annot=True
cbar=False
#plotting the heatmap
hm=sns.heatmap(data=data,linewidths=linewidths,linecolor=linecolor,annot=annot,cbar=cbar)
#displaying the plotted heatmap
plt.show()
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
print("The data to be plotted:\n")
print(data)
#setting the parameters values
linewidths=2
linecolor="yellow"
annot=True
cbar=False
xtickslabel=False
ytickslabel=False
#plotting the heatmap
hm=sns.heatmap(data=data,annot=annot,xticklabels=xtickslabel,yticklabels=ytickslabel,cbar=cbar)
#displaying the plotted heatmap
plt.show()
```
